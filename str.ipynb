```python
import requests
import json
from BeautifulSoup import BeautifulSoup, Comment
```

```python
api = 'http://edu.strbsu.ru/php/getList.php'
api_shedule = 'http://edu.strbsu.ru/php/getShedule.php'
```

```python
faculty = [4,5,6,7,8,9,10,11,18,26]

def parse_fac(resp):
    
    bs = BeautifulSoup(resp.text.encode('utf8'))
    
    results = {}
    
    for row in bs.findAll('a'):
        
        fid = row.attrs[1][1][15:-3]
        
        results[fid]=row.text
            
    return results    
 
    
    
results2 = {}

for f in faculty:
    
    print '************ ',f
    query = '?faculty='+str(f)

    full_query = api+query
        
    resp = requests.get(full_query)

    results2.update(parse_fac(resp))

len(results2)

# print (json.dumps(results2, indent=4, sort_keys=True, ensure_ascii=False))
```

```python
def extra_group(group_id):
    import dateutil.parser as dparser

    day ={}

    resp = requests.post(api_shedule,data={'type':'2','id':group_id,'week':'0'})
    bs = BeautifulSoup(resp.text.encode('utf8'))

    # удаляем теги комментов
    comments = bs.findAll(text=lambda text:isinstance(text, Comment))
    [comment.extract() for comment in comments]

    all = bs.findAll('div', 'day')

    # переводим текст в дату     
    lday=dparser.parse(all[0].h2.text,fuzzy=True).date()

    lessons={}
    lessons['lessons']={}

    for row in all[0].ul:

        if len(row)>2:
            less={}        
    #         for i in row:
    #             print i  

            number=int(row.find('div','number').text.strip('.'))

            if number:
                ltype=row.find('div','type').text
                less['type']=''.join([i for i in ltype if not i.isdigit()])
                less['id']=row.find('div','type').div.text
                less['cab']=row.find('div','cab').text
                less['time']=row.find('div','time').text
                less['name']=row.find('div','name').text
                less['prep']=row.find('div','prep').text

                lessons['lessons'][number]=less


    lessons['date']=lday
    lessons['date_name']=all[0].h2.text
    day[lday.strftime("%Y-%m-%d")]=lessons

#     print (json.dumps(day, indent=4, sort_keys=True, ensure_ascii=False, default=lambda x:str(x)))
    
    return day
    
extra_group(94)
```
